# Instalar Hugo

Instalamos Hugo de la manera recomendada por su página oficial con ayudad de Chocolatey.
### Instalación de Chocolatey
   1. Abrimos Power Shell como administrador, repito ABRIR COMO ADMINISTRADOR!.
   2. corre:

           Get-ExecutionPolicy
    
   - Si regresa:
    
            Restricted  
    
   - Ejecute:
   
           Set-ExecutionPolicy AllSigned
    
   - Otra opción es:
   
           Set-ExecutionPolicy Bypass -Scope Process
       
   3. Ahora ejecuta el siguiente comando:

           Set-ExecutionPolicy Bypass -Scope Process -Force; [System.Net.ServicePointManager]::SecurityProtocol = [System.Net.ServicePointManager]::SecurityProtocol -bor 3072; iex ((New-Object System.Net.WebClient).DownloadString('https://community.chocolatey.org/install.ps1'))
        
   4. Si no ve ningún error, ¡está listo para usar Chocolatey! Escriba:

            choco
    
   5. fuente oficial
    - https://chocolatey.org/install



### Instalación de Hugo
Ahora seguimos al pie de la letra la instalación en su página oficial.

   1. Abrimos Power Shell como administrador, repito ABRIR COMO ADMINISTRADOR!.
   2. Utilisa la version extendida para mayor covertura en cuanto a variedad de plantillas existen o van creando los     desarrolladores, corre el siguiente comando:
        
          choco install hugo-extended -confirm
          
   3. Si no ve ningún error, ¡está listo para usar Hugo! Escriba:

            hugo version
    
   4. fuente oficial
    - https://gohugo.io/getting-started/installing/
