# Instalación de Terminal Linux en Windows

Por favor instalar Terminal Linux para un mejor trabajo al editar sus páginas estáticas

#### Primero:

- Abrir inicio.
- Entrar en Configuración.
- Entrar en Actualización y seguridad.
- Entrar en Para Programadores.
- Activar Modo para desarrolladores.
- Clic en Sí:

![](../Paginas-Estaticas/acepta.PNG)



#### Segundo:

- Abrir inicio.
- Entrar en Configuración.
- Entrar en Aplicaciones.
- Entrar en Programas y características.

 

![acepta](aceptar2.png)


- Entrar en Activar o Desactivar las características de Windows.
- Marcar Subsistema de Windows para Linux y damos Aceptar.
![acepar3](aceptar3.png)

- Esperamos que cargue y damos clic en Reiniciar ahora.
- Ahora puedes usar Bash como terminal de Linux
- Crea tu usuario y confirma tu clave (***NO lo olvides, te servirá más adelante***), eso es todo!

![aceptar4](aceptar4.png)

### **Importante**

Si aun así no puedes usar bash has lo siguiente:

- Entra en Microsoft Store
    



- Descarga Ubuntu
    



- ahora solo abre la terminal de Ubuntu, crea tu usuario y confirma tu contraseña(***NO lo olvides, te servirá más adelante***)

- Tal vez exista un conflicto entre Ubuntu y bash puedes usar cualquiera

- Felicidades, ya tienes una terminal de Linux en tu Windows, eso es todo gracias por llegar hasta aquí :)


### Solución a errores

Si en algún momento no puedes usar la terminal de bash o Ubuntu.
- Ejecuta Power Shell como administrador



- Escribe el siguiente comando:
    - Enable-WindowsOptionalFeature -Online -FeatureName Microsoft-Windows-Subsystem-Linux
